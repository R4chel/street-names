```python
def get_street_names(line):
    line = line.lower()
    if line[50:51] == 's':
        return parse_s_type(line)
    else:
        return parse_non_s_type(line)
```

```python
def parse_s_type(line):
    stname = line[2:34].strip()
    num_of_progens = int(line[53:54])
    progen_1 = line[55:56] + ' ' + stname
    if num_of_progens == 1:
        return [progen_1]
    progen_2 = line[70:71] + ' '+ stname
    return [progen_1, progen_2]
```

```python
def parse_non_s_type(line):
    full_name = line[53:85].strip()
    stname = line[2:34].strip()
    spv = line[44:47]
    return [full_name]
```

```python
def synonyms(word):
    if word in abrev_dict:
        return abrev_dict[word].values
    return [word]
```

```python
def split_name(name):
    return map(synonyms, name.split())
```

```python
def possibilities(name):
    return get_strs(split_name(name))
```

```python
def tmp_possibilities(name):
    return map(lambda x: (x, name), possibilities(name))
```

```python
def add_word(word, l):
    return [x + [word] for x in l]
def flatten(l):
    return [x for y in l for x in y]
def helper(words, l):
    return flatten([add_word(word, l) for word in words])
def gen_strs(lol):
    return reduce(lambda a, x: helper(x, a), lol, [[]])
def get_strs(lol):
    return map(lambda x: ''.join(x) , gen_strs(lol))
```

```python
class Synonyms(object):
    def __init__(self, values):
        self.values = values
    
```

```python
abfile = open('data/streetab/abbrevs.txt', 'r')
abrevs = abfile.readlines()
```

```python
abrev_dict = {}
for ab in abrevs:
    words = ab.replace('\n','').lower().split()
    if words[0] == words[1]:
        continue
    syn = Synonyms(words)
    for word in words:
        abrev_dict[word] = syn
```
